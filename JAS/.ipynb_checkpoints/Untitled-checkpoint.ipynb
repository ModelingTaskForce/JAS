```python
import numpy as np
import pandas as pd
from models_b import SIR_gd
import matplotlib.pyplot as plt
```

```python
#Load data
data = pd.read_csv("../data/cases-brazil-cities-time.csv")
```

```python
#Selec RO state
data = data[data["state"] == "RO"]

#Put the date variable in the correct format
data["date"] = pd.to_datetime(data["date"])
```

```python
#Create the series for
data_grouped = data.groupby(["date"])["deaths", "totalCases"].sum().reset_index()
data_grouped["day"] = range(1,len(data_grouped["totalCases"]) + 1)
```

```python
sir = SIR_gd(pop = 1749000)
```

```python
sir.fit(x = data_grouped["date"],
        y = data_grouped["totalCases"],
        n_tries = 30)
```

```python
ypred = sir.predict(time = data_grouped["date"])
```

```python
fig, ax = plt.subplots(nrows = 1, ncols = 1)
ax.plot(data_grouped["date"], data_grouped["totalCases"], "ro")
ax.plot(data_grouped["date"], np.cumsum(ypred))
plt.savefig("Sir_trad.png")
```

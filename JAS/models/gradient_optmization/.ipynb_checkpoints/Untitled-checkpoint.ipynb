```python
import sir
import pandas as pd
import matplotlib.pyplot as plt
import altair as alt
plt.style.use('seaborn-talk') # beautify the plots!
```

```python
#Load data
data = pd.read_csv("https://raw.githubusercontent.com/wcota/covid19br/master/cases-brazil-states.csv")
data.head()
```

```python
#Selec RO state
data = data[data["state"] == "RO"]

#Put the date variable in the correct format
data["date"] = pd.to_datetime(data["date"])
data["day"] = range(1,len(data["totalCases"]) + 1)
data.head()
```

```python
confirmed_plot = alt.Chart(data).mark_circle(size = 60, color =' blue').encode(
    x = alt.X('date', axis=alt.Axis(title = 'Date')),
    y = alt.Y('totalCases', axis=alt.Axis(title = 'Cases'))
)

deaths_plot = alt.Chart(data).mark_circle(size = 60, color = 'red').encode(
    x='date',
    y='deaths'
)
state_graph =  confirmed_plot + deaths_plot
state_graph.interactive()
```

```python
model = sir.start_model(pop = 1749000)
```

```python
model.fit(x = data["day"], y = data["totalCases"], n_tries = 20, n_betas = 3)
```

```python
model.get_parameters()
```

```python
time = pd.date_range(start = data["date"].iloc[0], end = pd.to_datetime("2020-12-31"))
series = model.predict(time = time)
```

```python
data["predicted"] = series["Tt"][0:len(data["totalCases"])]
data.head()
```

```python
predicted_plot = alt.Chart(data).mark_circle(size = 60, color = 'red').encode(
    x = 'date',
    y = 'predicted'
)
fit_graph =  confirmed_plot + predicted_plot 
fit_graph.interactive()
```

```python
_, ax = plt.subplots(nrows = 1, ncols = 1, figsize = [28,20])
ax.plot(time, series["S"], "g")
ax.plot(time, series["I"], "r")
ax.plot(time, series["R"], "b")
```
